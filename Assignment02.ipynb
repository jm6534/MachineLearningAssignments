```python
import numpy as np
import matplotlib.pyplot as pp
import math
```

define differentiable Function :

$$ f(x) = {1 \over 9}x^3 $$

```python
def origFunction(x):
    return 1 / 9 * np.power(x,3)
```

define a domain of function

```python
domain = np.arange(-5, 5, 0.1)
```

plot the function

```python
pp.plot(domain, origFunction(domain), 'black')
pp.show()
```

select a point within the domain

```python
x = 3
```

mark the seleted point on the function

```python
mark = [81]
pp.plot(domain, origFunction(domain), color='black', marker='D', markevery=mark)
pp.show()
```

define derivative origFunction : 

$$ f'(x) = {1 \over 3}x^2 $$

```python
def derFunction(x):
    return 1 / 3 * np.power(x,2)
```

define the first-order Taylor approximation at the selected point

```python
def taylor(x,point):
    return origFunction(point)+derFunction(point)*(x-point)
```

plot the taylor approximation with the same domain of the original function

```python
pp.plot(domain, taylor(domain,3), color='green')
pp.plot(domain, origFunction(domain), color='black', marker='D', markevery=mark)
pp.show()
```
